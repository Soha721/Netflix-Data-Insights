```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
print(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.dropna(subset=['rating','director','cast','country','date_added'])
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
type_counts = df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values ,color = ['skyblue','orange'])
plt.title('Number of movies vs Tv shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('Number of movies vs Tv Shows on Netflix', dpi = 300,bbox_inches = 'tight')
plt.show()
```

```python
rating_counts = df['rating'].value_counts()
plt.figure(figsize=(4,8))
plt.pie(rating_counts, labels = rating_counts.index, autopct = '%1.1f%%',startangle=90)
plt.title('Rating counts')
plt.tight_layout()
plt.savefig('Rating_count.png', dpi = 300,bbox_inches = 'tight')
plt.show()
```

```python
release= df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.plot(release.index, release.values,color='red')
plt.title('Release count of movies')
plt.xlabel('Release Year')
plt.ylabel('Release Count')
plt.tight_layout()
plt.savefig('Release count of movies.png', dpi=300)
```

```python
movie=df[df['type'] == 'Movie'].copy()
movie['duration_int'] = movie['duration'].str.replace('min','').astype(int)
plt.hist(movie['duration_int'],bins = 30,color ='purple', edgecolor='black')
plt.title('Distribution of Movie Duration')
plt.xlabel('Duration of movie in minutes')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.savefig('movie_duration_histogram.png')
plt.show()
```

```python
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(4,8))
plt.barh(country_counts.index, country_counts.values, color='teal')
plt.title('top 10 countries by number of shows')
plt.xlabel('Number of shows')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('top10_countries.png')
plt.show()
```

```python
content_by_year = df.groupby(['release_year', 'type']).size().unstack().fillna(0)

fig, ax = plt.subplots(1, 2, figsize=(12,6))  # Make it wider

# Plot Movies
ax[0].plot(content_by_year.index, content_by_year['Movie'], color='teal')
ax[0].set_title('Movies Released per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number of Movies')

# Plot TV Shows
ax[1].plot(content_by_year.index, content_by_year['TV Show'], color='orange')
ax[1].set_title('TV Shows Released per Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Number of TV Shows')
plt.savefig('Movies Released per Year.png')
plt.tight_layout()
plt.show()
```
